Merging AND Concat

```python
import pandas as pd
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4,5],"B":[11,22,33,44,55]})
var2 = pd.DataFrame({"A":[1,2,3,4,5],"C":[99,88,77,66,65]})
```

```python
pd.merge(var1,var2)
```

```python
pd.merge(var2,var1)
```

```python
pd.merge(var1,var2,how="left")
```

```python
pd.merge(var1,var2,how="right")
```

```python
pd.merge(var1,var2,how="outer",indicator=True)
```

```python
var3 = pd.DataFrame({"A":[1,2,3,4,5],"B":[11,22,33,44,55]})
var4 = pd.DataFrame({"A":[1,2,3,4,5],"B":[99,88,77,66,65]})
```

```python
pd.merge(var3,var4,left_index=True,right_index=True,suffixes=("Ajmayen","Python"))
```

Concat

```python
sr1 = pd.Series([1,2,3,4,5])
sr2 = pd.Series([22,33,44,55,66])
pd.concat([sr1,sr2])
```

```python
d1 = pd.DataFrame({"A":[1,2,3,4,5],"B":[11,22,33,44,55]})
d2 = pd.DataFrame({"A":[1,2,3,4,5],"B":[99,88,77,66,65]})
pd.concat([d1,d2],axis=1)
```

```python
d1 = pd.DataFrame({"A":[1,2,3,4,5],"B":[11,22,33,44,55]})
d2 = pd.DataFrame({"A":[1,2,3,4,5],"B":[99,88,77,66,65]})
pd.concat([d1,d2])
```

```python
d3 = pd.DataFrame({"A":[1,2,3,4,5],"B":[11,22,33,44,55]})
d4 = pd.DataFrame({"A":[3,4,5],"B":[77,66,65]})
pd.concat([d3,d4],axis=1,join="inner")
```

```python
d1 = pd.DataFrame({"A":[1,2,3,4,5],"B":[11,22,33,44,55]})
d2 = pd.DataFrame({"A":[1,2,3,4,5],"B":[99,88,77,66,65]})
pd.concat([d1,d2],axis=0,keys=["d1","d2"])
```

```python
d1 = pd.DataFrame({"A":[1,2,3,4,5],"B":[11,22,33,44,55]})
d2 = pd.DataFrame({"A":[1,2,3,4,5],"B":[99,88,77,66,65]})
pd.concat([d1,d2],axis=1,keys=["d1","d2"])
```

```python
d1 = pd.DataFrame({"A":[1,2,3,4,5]})
d2 = pd.DataFrame({"A":[1,2,3,4,5],"B":[99,88,77,66,65]})
pd.concat([d1,d2])
```
